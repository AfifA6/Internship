Q.1 & 2. Scraping Guitar Data from amazon.in 

```python
#Importing necessary libraries
import pandas as pd
import selenium
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from bs4 import BeautifulSoup
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
#Opening web browser
driver = webdriver.Chrome(r"C:\Users\skabd\Downloads\chromedriver.exe")

#Opening amazon.com page
driver.get('https://www.amazon.in/')
```

```python
#Entering product name in search box
product = driver.find_element(By.XPATH, '/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input')
product.send_keys('Guitar')
```

```python
#Clicking search button
search_button = driver.find_element(By.XPATH, '//div[@class="nav-search-submit nav-sprite"]//span')
search_button.click()
```

```python
#Fetching url to open each product
product_opening_url = []

start = 0
end = 3

for page in range(start, end):
    url = driver.find_elements(By.XPATH, '//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        product_opening_url.append(i.get_attribute("href"))
    next_button = driver.find_element(By.XPATH, '//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]') 
    
    next_button.click()
    time.sleep(2)
```

```python
len(product_opening_url)
```

```python
product_opening_url
```

```python
#Clicking on first product
first_product = driver.find_element(By.XPATH, '/html/body/div[1]/div[2]/div[1]/div[1]/div/span[1]/div[1]/div[3]/div/div/div/div/div/div/div[2]/div[1]/h2/a/span')
first_product.click()
```

```python
#Creating empty lists
Brand = []
Product = []
Price = []
Return_Exchange = []
Delivery = []
Availability = []
Product_URL = []
```

```python
for url in product_opening_url:
    driver.get(url)
    time.sleep(2)
    
    try:
        brand = driver.find_element(By.XPATH, '//*[@id="productOverview_feature_div"]/div/table/tbody/tr[1]/td[2]/span')
        Brand.append(brand.text)
        
    except NoSuchElementException:
        Brand.append('-')
        
    #for product name
    try:
        product = driver.find_element(By.XPATH, '//*[@id="productOverview_feature_div"]/div/table/tbody/tr[1]/td[3]/span')
        Product.append(product.text)
        
    except NoSuchElementException:
        Product.append('-')
        
        
    #for price
    try:
        price = driver.find_element(By.XPATH, '//*[@id="corePriceDisplay_desktop_feature_div"]/div[1]/span[2]/span[2]/span[2]')
        Price.append(price.text)
        
    except NoSuchElementException:
        Price.append('-')
        
        
    #for Return Exchange
    try: 
        ret_ex = driver.find_element(By.XPATH, '//*[@id="RETURNS_POLICY"]/span/div[2]/a')
        Return_Exchange.append(ret_ex.text)
        
    except NoSuchElementException:
        Return_Exchange.append('-')
        
     #for Delivery
    try:
        delivery = driver.find_element(By.XPATH, '//*[@id="mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE"]/span/a[1]')
        Delivery.append(delivery.text)
        
    except NoSuchElementException:
        Delivery.append('-')
        
     #for availability
    try:
        availability = driver.find_element(By.XPATH, '//*[@id="availability"]/span')
        Availability.append(availability.text)
        
    except NoSuchElementException:
        Availability.append('-')
        
     
    #for product url
    try:
        product_url = driver.find_element(By.XPATH, '//*[@id="search"]/div[1]/div[1]/div/span[1]/div[1]/div[3]/div/div/div/div/div/div/div[2]/div[1]/h2/a')
        Product_URL.append(product_url.get_attribute('href'))
        
    except NoSuchElementException:
        Product_URL.append('-')
        
    
        
        
```

```python
print(len(Brand), len(Product), len(Price), len(Return_Exchange), len(Delivery), len(Availability), len(Product_URL))
```

```python
df1 = pd.DataFrame({'Brand_Name': Brand, 'Product_Name': Product, 'Price': Price, 'Return/Exchnage': Return_Exchange, 'Delivery': Delivery, 'Availability': Availability, 'Product_URL': Product_URL})
df1
```

```python
df1.to_csv("Guitar_data.csv")
```

```python
driver.close()
```
